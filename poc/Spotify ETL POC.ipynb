#### Some date/time manipulation

```python
import datetime
from pytz import timezone
test_df['date_played'] = [datetime.datetime.strptime(date_time_str, '%Y-%m-%dT%H:%M:%S.%fZ') for date_time_str in test_df['time_played']]
test_df['date_played'] = [d.date() for d in test_df['date_played']]
test_df['c_time_played'] = [datetime.datetime.strptime(date_time_str, '%Y-%m-%dT%H:%M:%S.%fZ') for date_time_str in test_df['time_played']]
test_df['c_time_played'] = [timezone('UTC').localize(d) for d in test_df['c_time_played']]
test_df['c_time_played'] = [d.astimezone(timezone('US/Central')) for d in test_df['c_time_played']]
test_df['c_time_played'] = [d.time() for d in test_df['c_time_played']]
```
